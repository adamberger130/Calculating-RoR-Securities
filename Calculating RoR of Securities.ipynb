```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
PG = pd.read_csv('/Users/adamberger/Desktop/PG.csv', index_col = 'Date')
```

```python
PG['simple_return'] = (PG['Adj Close'] / PG['Adj Close'].shift(1)) - 1
print(PG['simple_return'])
```

```python
PG['simple_return'].plot(figsize=(8,5))
```

```python
avg_returns_a = PG['simple_return'].mean()*250
avg_returns_a
```

```python
print(str(round(avg_returns_a, 5)* 100) + '%')
```

```python
PG['log_returns'] = np.log(PG['Adj Close'] / PG['Adj Close'].shift(1))
PG['log_returns']
```

```python
PG['log_returns'].plot(figsize=(8,5))
```

```python
log_returns_a = PG['log_returns'].mean()*250
log_returns_a
```

```python
print(str(round(log_returns_a, 5)* 100) + '%')
```
